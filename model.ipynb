```python

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pickle
import seaborn as sns

```

```python
# Load the dataset
df = pd.read_excel('/content/recipees.xlsx')



df

```

```python
df.describe()

df.shape

```

```python
sns.pairplot(df[['calories', 'protein', 'carbs', 'fat', 'price($)']])

```

```python
import numpy as np

#Correlation Matrix
sns.heatmap(df[['calories', 'protein', 'carbs', 'fat', 'price($)','category']].corr(), annot=True, cmap='coolwarm')


```

```python
# Features for clustering
X = df[['calories', 'protein', 'carbs', 'fat', 'price($)']]



X.shape


```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)



```

```python
X_scaled
```

```python

# Train a K-Means model
n_clusters = 4  # Number of clusters can be adjusted
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X_scaled)



```

```python

df['Cluster'] = kmeans.labels_
```

```python
# Calculate the silhouette score
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X_scaled, df['Cluster'])
print(f'Silhouette Score: {silhouette_avg}')
```

```python
# Save the model and scaler to .sav files
with open('kmeans_model11.sav', 'wb') as model_file:
    pickle.dump(kmeans, model_file)
with open('scaler_model11.sav', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)

print("Model and scaler saved successfully.")


```

### **Extra**

### **Hierarchical Clustering**


```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Perform hierarchical clustering
Z = linkage(X_scaled, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.show()

# Get cluster labels
cluster_labels = fcluster(Z, n_clusters, criterion='maxclust')
```

```python
#Silhouette Score for heirarchical clustering
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X_scaled, cluster_labels)
print(f'Silhouette Score: {silhouette_avg}')

```
